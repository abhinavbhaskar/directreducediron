1. 100% DRI is fed into the furnace without any scrap. The quality of scrap has a signifcant effect on 
the energy consumption in a DRI. Since its difficult to find the exact compostion it has been considered that 
all the input material will be DRI directly fed from the direction reduction chamber within the shaft.
2. Hot DRI is fed into the DRI at 600 C as it saves considerable amount of electrical energy.
3. Natural gas is not used for heating the material as its the general practice to use natural gas with 
scrap for initial heating.
4. As DRI is reduced only with Hydrogen, it is assumed that it does not contain any Fe3C
5. Carbon is added into the EAF to reduce the remaining FeO in the mix and also to generate CO for froth formation 
which is essential for the operation of the EAF and to extend the life of the graphite electrodes and the refractory.
6. CaO and MgO are added in the EAF as slag formers to maintain the basicity of the EAF.

9. The weight of CaO and MgO used are according to data published in the literature.
10. Latent heat of fusion is added in the enthalpy calculations for Fe and FeO 
11. The model is a steady state simulation and does not considere the time dependency of different parametes.
12. Time etaken for a 150 Ton EAF charge to finish melting is 45 minutes in a typical industrial furnace.The temperature of input DRI has a direct correlation with the time taken 
for the process.

## Energy balance of the EAF system with DRI
 
The modelling work is baseed on the  https://www.sciencedirect.com/science/article/pii/S036054421100524X. 


 **_EAF energy balance  - Equation 1_**


$\begin{equation} \label{Energy balance in EAF}\large E_{total}=E_{in}=\large \int_{\small charging}^{\small tapping} P_{\small electric} dt + \Delta H_{\small oxygen injection} + \Delta H_{\small ng_burners} = E_{out}=\Delta H_{\small steel }+ \Delta H_{\small slag } + \large\int\Delta H^._{\small off-gas}dt +\large\int\Delta Q^._{\small cooling}dt+\large\int\Delta Q^._{\small radiaitve losses}dt \end{equation}$




The Enthalpy that is heating up the melt and slag in a small time step $\Delta t$ is determined from the energy input from the electric arc , ($P_{electric}$), from oxidation reactions, and from the gas burners ($P_{ngburners})$. Equation 2 depicts the same in mathematical form : 

**_Energy balance -Equation 2_**


$\Delta E_{Total}=\large \int_{\small t}^{\small t+\Delta t}P_{elec}+ \int_{\small t}^{\small t+\Delta t}\sum_{i}^{n}m^._{oxidized,i}h_{oxidized,i}dt +\large \int_{\small t}^{\small t+\Delta t}P_{ng_burners}dt=\large \int_{\small T(t)}^{\small T( t+\Delta t)} h_{steel}m_{steel}dt + \large \int_{\small T(t)}^{\small T( t+\Delta t)}h_{slag}m_{slag}dt+\large \int_{\small t}^{\small t+\Delta t}P_{off-gas}dt+\large \int_{\small t}^{\small t+\Delta t}P_{cooling}dt + \large \int_{\small t}^{\small t+\Delta t}P_{other losses}dt$




Efficiency values for energy conversion for electric arc furnaces 

**_Electric arc -Equation 3_**


$\Delta H_{melt,el}=\eta_{el}\eta_{arc}\Large \int_{\small t}^{\small t+\Delta t}P_{elec}dt$


In this model , the value of $\eta_{el}\eta_{arc}$ has been taken to be 0.6 as the melting rate is lower for 100% DRI than with scrap based mixture. 

Enthalpy portion that is transferred into the melt and the slag from the oxidation reactions which are exothermic. 

**_Chemical enthalpy -Equation 4_**



$\Delta H_{melt,chem}=\eta_{chem}\Large\int_{\small t}^{\small t+\Delta t}\sum_{i}^{n}m^._{oxidized,i}h_{oxidized,i}dt$


$\eta_{chem}$ is considered to be 0.5 to 0.6 in this work.



Enthalpy transferred from the natural gas burners in the EAF is calculated as : 

**_Natural gas burners calculations -Equation 5_**



$\Delta H_{melt,burners}=\eta_{burners}\Large \int_{\small t}^{\small t+\Delta t}P_{ng_burners}dt$



In the model, energy input from natural gas has not been considered as it is primarily used with scrap based EAF systems to provide initial heating and reducing the overall energy consumption of the EAF. 



The terms quantifying the $\Delta H_{\small ng_burners},\Delta H^._{\small off-gas},\Delta Q^._{\small cooling},\Delta Q^._{\small radiaitve losses}  $ in Equation 1 can be ignored. The **losses are considered in the efficiency factors of the electric arc and  chemical enthalpy calculations**. Natural gas burners are not used for the 100 % DRI based EAF systems.  

Final expression of the energy balance can be represented by **_Equation : 6_**



$\Delta H_{melt}=\eta_{el}\eta_{arc}\int_{ t}^{t+\Delta t}P_{elec}dt + \eta_{chem}\int_{ t}^{ t+\Delta t}\sum_{i}^{n}m_{oxidized,i}h_{oxidized,i}dt = \int_{T(t)}^{T( t+\Delta t)} h_{steel}m_{steel}dt + \int_{T(t)}^{ T( t+\Delta t)}h_{slag}m_{slag}dt $
